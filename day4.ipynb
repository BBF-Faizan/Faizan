```python
import pyttsx3 as pytt
```

```python
ttsx=pytt.init()
```

```python
voices = ttsx.getProperty('voices')
for voice in voices:
   ttsx.setProperty('voice', voice.id)
```

```python
# voice ka gender change kane liye hai 
```

```python
ttsx.say(" hello every one i am going explain about pyttsx3")
```

```python
ttsx.runAndWait()
```
